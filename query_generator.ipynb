```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("vicky7901/my_LLaMA-2-model")
model = AutoModelForCausalLM.from_pretrained("vicky7901/my_LLaMA-2-model")
```

```python
from datasets import load_dataset


def get_input_from_user():
  """Gets input from the user."""
  prompt = input("Enter the input: ")
  return prompt

# Get input from the user
user_input = get_input_from_user()


# Create the prompt
prompt = f"""### Instruction:
Generate all possible questions about the columns names provided in the input.

### Input:
{user_input}

### Response:
"""


# Convert the prompt to a tensor
input_ids = tokenizer(prompt, return_tensors="pt", truncation=True).input_ids


# Generate the output
# with torch.inference_mode():
outputs = model.generate(input_ids=input_ids, max_new_tokens=100, do_sample=True, top_p=0.9, temperature=0.9)


# Print the results
print(f"Prompt:\n{user_input}\n")
print(f"Queries:\n{tokenizer.batch_decode(outputs.detach().cpu().numpy(), skip_special_tokens=True)[0][len(prompt):]}")
#print(f"Ground truth:\n{sample['### Response']}")
```
